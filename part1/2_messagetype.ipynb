```python

from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)

from langchain_core.prompts import PromptTemplate

dynamic_prompt = PromptTemplate.from_template("say something interesting about {topic}?")

user_input = input("Enter a topic: ")

ready_prompt = dynamic_prompt.invoke({"topic": user_input})

llm_gemini.invoke(ready_prompt)
```

```python
from langchain_core.prompts import ChatPromptTemplate

promt_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a {role}."),
        ("user", "say something interesting about {topic}?")
    ]
)

ready_pront = promt_template.invoke({"topic": "dinasaures", "role": "funny friend"})
print(ready_pront)
answer = llm_gemini.invoke(ready_pront)
print(answer.content)
```
