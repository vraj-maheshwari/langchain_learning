```python
from langchain_core.messages import HumanMessage, SystemMessage, AIMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
```

# **using instruction in promt**

```python
answer = llm_gemini.invoke("Tell dinosaur facts. Return output format is key value pair as 'title' and 'detail' and do not use markdown")
answer.content
```

# **using pydantic**

```python
from pydantic import BaseModel,Field

class llm_Schema(BaseModel):
    title: str = Field(description="the title of the facts")
    detail: str = Field(description="the detail of the facts")
```

```python
# obj = llm_Schema(title="T-Rex", detail="The T-Rex was a large carnivorous dinosaur.")
obj = llm_Schema(name="T-Rex", detail="The T-Rex was a large carnivorous dinosaur.")
obj
```

```python
llm_gemini_with_structured_output = llm_gemini.with_structured_output(llm_Schema)
answer = llm_gemini_with_structured_output.invoke("Tell dinosaur facts")
answer
```

```python
from typing import TypedDict

class llm_TypedDict(TypedDict):
    title: str
    detail: str
```
