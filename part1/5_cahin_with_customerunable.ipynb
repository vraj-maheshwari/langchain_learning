# promt -> LLM -> parser ->cutome function who convert answer into dictonry ->promt -> LLM -> Parser

# **1 ai call ->custome fucntion ->2 ai call**

```python
#step 1 promt
from langchain_core.prompts import ChatPromptTemplate

prompt_text = ChatPromptTemplate.from_messages([
    ('system', 'you are a helpful assistant'),
    ('human', '{input}')
])
```

```python
#step2 LLM
from langchain_google_genai import ChatGoogleGenerativeAI

llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
```

```python
#step 3 tring parser 
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

```python
#teps 4 custome runnable function
from langchain_core.runnables import RunnableLambda

def dictinary_maker(text: str) -> dict:
    return {"text": text}

runnable_dictonary_maker = RunnableLambda(dictinary_maker)
```

```python
#step 5 promt 2
from langchain_core.prompts import ChatPromptTemplate

prompt_post = ChatPromptTemplate.from_messages([
    ('system', 'you are a social media post generator'),
    ('human', 'create the post for the following text for linkdin :{text}')
])
```

```python
#steps 6 LLM 2
from langchain_google_genai import ChatGoogleGenerativeAI
llm_gemini_2 = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
```

```python
#stesp 7 output string parser
str_parser_2 = StrOutputParser()
```

```python
#step 8 chain

chain = prompt_text | llm_gemini | str_parser | runnable_dictonary_maker | prompt_post | llm_gemini_2 | str_parser_2
answer = chain.invoke("What is the capital of France?")
answer
```
