### **2.1 chain**

```python
#chain 2.1
def insta_chain(text:dict):

    text = text['text']

    #step 1 promt
    from langchain_core.prompts import ChatPromptTemplate

    prompt_instagram = ChatPromptTemplate.from_messages([
        ('system', 'you are instagram post genrator'),
        ('human', 'create post for following text for instagram: {text}')
    ])

    #step2 LLM
    from langchain_google_genai import ChatGoogleGenerativeAI

    llm_gemini2 = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)


    #step 3 tring parser 
    str_parser = StrOutputParser()

    chain_insta = prompt_instagram | llm_gemini2 | str_parser

    result = chain_insta.invoke(text)

    return result
```
