```python
#step 1 promt
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ('system', 'you are a helpful assistant'),
    ('human', '{input}')
])
```

```python
#step2 LLM
from langchain_google_genai import ChatGoogleGenerativeAI

llm_gemini = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0)
```

```python
#step 3 tring parser 
from langchain_core.output_parsers import StrOutputParser

str_parser = StrOutputParser()
```

# **chain example**

```python
chain = prompt | llm_gemini | str_parser
answer = chain.invoke("Who is pm of india?")
answer
```

# **chain example2**

```python
from langchain_core.runnables import RunnableSequence
chain2 = RunnableSequence(prompt, llm_gemini, str_parser)
answer = chain2.invoke("Who is pm of india?")
answer
```
